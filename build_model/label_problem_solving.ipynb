### 이미지와 라벨 맞지 않는 경우 해결

```python
import os
a = os.listdir('dataset/images')
a = sorted(a)
print(a)
```

```python
import os
b = os.listdir('dataset/labels')
b = sorted(b)
print(b)
```

```python
namelist = []
problem = []

for i in range(len(a)):
    example1 = a[i]
    example2 = b[i]
    name1, ext = os.path.splitext(example1)
    name2, ext = os.path.splitext(example2)

    if name1 == name2:
        namelist.append(name1)
    else:
        problem.append(name1)

print(len(namelist))
#problem에 문제 있으면 이미지 인덱스로 뜸. 그 부분 수정하면됨.
print(len(problem))
```
