## 4. Real-time detection (Pretrained example)

```python
cap = cv2.VideoCapture(0) #영상은 같은 directory에 놓고 0 대신 'library1.mp4'이런식으로 넣으면 됨. 0은 내 웹캠!
while cap.isOpened():
    ret, frame = cap.read()
    
    # detection
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```

## 6. 학습시킨 모델 load

위에서 학습 시켰으면 'yolov5/runs/train//*exp(알맞은폴더)*/'에 weights 폴더가 생성됐을 것. 이를 바탕으로 아래의 코드 작성

```python
model = torch.hub.load('ultralytics/yolov5','custom', path='yolov5/runs/train//*exp(알맞은폴더)*//weights/last.pt', force_reload=True)
```

```python
cap = cv2.VideoCapture(0) #영상은 같은 directory에 놓고 0 대신 'library1.mp4'이런식으로 넣으면 됨. 0은 내 웹캠!
while cap.isOpened():
    ret, frame = cap.read()
    
    # detection
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
