## 2. load pretrained model

```python
#download pretrained model for torch hub
model = torch.hub.load('ultralytics/yolov5','yolov5s')
```

## 4. Real-time detection (Pretrained example)

```python
cap = cv2.VideoCapture(0) #  directory  0  'library1.mp4'  . 0  !
while cap.isOpened():
    ret, frame = cap.read()
    
    # detection
    results = model(frame)
    
    cv2.imshow('YOLO', np.squeeze(results.render()))
    
    if cv2.waitKey(10) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
